combining DictVectorizer and mapper.

Note: This transformer will only do a binary one-hot encoding when feature values are of type string. If categorical features are represented as numeric values such as int, the DictVectorizer can be followed by OneHotEncoder to complete binary one-hot encoding.

An easier way would just be to convert categorical integers to objects

```python
dvec = DictVectorizer(sparse=False)
```

```python
X_dict = dvec.fit_transform(base.iloc[:, 2:].transpose().to_dict().values())
```

```python
X = pd.DataFrame(X_dict, columns=dvec.get_feature_names(), index=base.index)
```
