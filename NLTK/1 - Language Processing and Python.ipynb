```python
# Frequency distributions
fdist1 = nltk.FreqDist(book.text1)
fdist1 # a frequency distribution
```

```python
fdist1.most_common(50) # gives the most common words
```

```python
fdist1.hapaxes() # hapaxes are words that appear only once
```

```python
# Fine-grained selection of words
# to look at long words
V = set(book.text1)
long_words = [w for w in V if len(w) > 15] # for each word w in vocabulary V, check if len(w) is greater than 15
sorted(long_words)
```

```python
# it might be better to find frequently occuring long words
fdist5 = nltk.FreqDist(book.text5)
sorted(w for w in set(book.text5) if len(w) > 7 and fdist5[w] > 7) 
```

```python
# collocations and bigrams
# collocation is a series of words that occur together unusually often e.g. red wine is a collocation but the wine is not
list(nltk.bigrams(['more', 'is', 'said', 'than', 'done']))
```

```python
# Now, collocations are essentially just frequent bigrams, except that we want to pay more attention to the cases that involve 
# rare words. In particular, we want to find bigrams that occur more often than we would expect based on the frequency of the 
# individual words. The collocations() function does this for us.
book.text4.collocations()
book.text8.collocations()
```

```python
# we can also count word lengths
[len(w) for w in book.text1]
fdist = nltk.FreqDist([len(w) for w in book.text1])
```

```python
fdist 
fdist.most_common()
fdist.max()
fdist[3]
fdist.freq(3)
type(fdist)
```
