```python
# Import the Pandas library
import pandas as pd

# Load the train and test datasets to create two DataFrames
train_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
train = pd.read_csv(train_url)

test_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv"
test = pd.read_csv(test_url)
#Print the `head` of the train and test dataframes
print(train.head())
print(test.head())

train.describe() # Same as summary in R
train.shape # same as dim
```

```python
# Absolute numbers
train["Survived"].value_counts() # survived is the column name and value_counts() works like table() in R
# Percentages
train["Survived"].value_counts(normalize = True)

# We can perform similar counts and perentages on subsets of the Survived column
# Let's see if gender has a role in survival

train["Survived"][train["Sex"] == "male"].value_counts()
train["Survived"][train["Sex"] == "female"].value_counts()
# Female survival ratio is much higher than male 
# these can also be looked at as proportions by using normalize = True
```

```python
# Another variable that can influence survival is age as it is probable that children were saved first. We can create a new 
# categorical variable names child. We create a new column and then assign it a value. 
train["Child"] = float('NaN') # creating a new 
train["Child"][train["Age"] < 18] = 1
train["Child"][train["Age"] >= 18] = 0
print(train["Child"])

print(train["Survived"][train["Child"] == 1].value_counts(normalize = True))
print(train["Survived"][train["Child"] == 0].value_counts(normalize = True))

```

```python
# We saw that the chances of females surviving is higher than males
# test the prediction that all females survive and all males die
test_one = test
test_one["Survived"] = 0
test_one["Survived"][test_one["Sex"] == "female"] = 1
print(test_one["Survived"])
```

```python
# Cleaning the date before creating the tree
# We will replace all missing age values with the median
train["Age"] = train["Age"].fillna(train["Age"].median())

# Another problen is that Sex and Embarked are categorical but non numeric and python cannot handle that (that sucks)
# Assign integer 1 to females
train["Sex"][train["Sex"] == "female"] = 1
train["Sex"][train["Sex"] == "male"] = 0

# Embarked has some missing values that we will impute as "S" (the most common variable)
train["Embarked"] = train["Embarked"].fillna("S")
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

print(train[["Sex", "Embarked"]])
```
